```python
import pandas as pd
import os
data_path = "data_sample_clpsych19/"
```

```python
def json_creator(data_frame,user_category):
    data_in_json = []
    
    user_id = int(data_frame['user_id'][0])
    
    label = user_category
    posts = []
    user = {}
    
    for index,row in data_frame.iterrows():
        one_post = []
        one_post.append(row['post_id'])
        one_post.append(row['timestamp'])
        one_post.append(row['subreddit'])
        one_post.append(row['post_title'])
        one_post.append(row['post_body'])

        
        posts.append(one_post)
    
    user ['user_id'] = user_id
    user ['label'] = label
    user ['posts'] = posts
    
    return user
```

```python
all_user_files = [f for f in os.listdir(data_path) if not f.startswith('.')]
data_in_json = []

for file in all_user_files:
    
    if '-' in file:
        user_cat = "Control"
    else:
        user_cat = "SucideWatch"

    
    df = pd.read_csv(data_path+file)
    data_in_json.append(json_creator(df,user_cat))
```

```python
print(len(data_in_json))
```

```python
import json
with open ('json_format_data.txt','w') as outfile:
    #for user in data_in_json:
    json.dump(data_in_json,outfile)
    outfile.write('\n')
```
